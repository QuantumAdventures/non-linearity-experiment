```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import seaborn as sns
from tqdm import tqdm
```

```python
sig = np.random.randn(300000)
t = np.linspace(0, 0.25, sig.shape[0])
fs = 1/np.diff(t)[0]
```

```python
sos = scipy.signal.butter(1, 1, 'lp', fs=fs, output='sos')
filtered = scipy.signal.sosfilt(sos, sig)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(t, sig)
ax1.set_title('10 Hz and 20 Hz sinusoids')
ax1.set_xlim([0, t[-1]])
ax2.plot(t, filtered)
ax2.set_title('After 15 Hz high-pass filter')
ax2.set_xlim([0, t[-1]])
ax2.set_xlabel('Time [seconds]')
plt.tight_layout()
plt.show()
```

```python
f_white, power_white = scipy.signal.periodogram(sig, fs=fs, window='hamming')
f_brown, power_brown = scipy.signal.periodogram(filtered, fs=fs, window='hamming')
```

```python
plt.plot(f_white, power_white)
plt.plot(f_brown, power_brown)

#plt.xlim([1e4, 5*1e5])
plt.grid()
plt.yscale('log')
plt.xscale('log')
```

```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[int(result.shape[0]/2):]
```

```python
corr = autocorr(filtered)
```

```python
#plt.plot(autocorr(sig))
plt.plot(corr)
plt.grid()
plt.show()
```
