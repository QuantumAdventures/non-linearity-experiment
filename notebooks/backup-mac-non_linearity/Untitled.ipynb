```python
import os
import matplotlib.pyplot as plt
import scipy.signal as signal
import pandas as pd


folders = ['12mbar - Sem Filtro',
 'Sem NP - Sem Filtro',
 'Sem NP - Com Filtro',
 '12mbar - Com Filtro']

windows = 10
channel = 'ch1'
welchMethod = 1
```

```python
def psd(path_folders, windows, channel, welchMethod):
    folders = next(os.walk(path_folders))[1] #don't modify this line!!
    selectedFolders = folders #if you want to ignore some folders, you can pass here a list of the desirable folders. Else, just leave selectedFolders = folders
    powerList = []
    for folder in folders:
        if folder in selectedFolders:
            path = os.path.join(path_folders,folder)
            files = next(os.walk(path))[2]
            data = []
            for file in files:
                if file.endswith('.pkl'):
                    filePath = os.path.join(path_folders,folder,file)
                    data.append(pd.read_pickle(filePath))
            dt = data[0].t[1]-data[0].t[0]
            f = 1/dt
            if welchMethod == 1: #if welch method is ON
                freq, power = signal.welch(data[0][channel], f, window = 'hamming', nperseg = int(len(data[0][channel])/windows))
                for i in range(1,len(data)):
                    freq, powerTemp = signal.welch(data[i][channel], f, window = 'hamming', nperseg = int(len(data[i][channel])/windows))
                    power += powerTemp
            else: #if welch method is OFF
                freq, power = signal.periodogram(data[0][channel], f, scaling='density')
                for i in range(1,len(data)):
                    freq, powerTemp = signal.periodogram(data[i], f, scaling='density')
                    power += powerTemp
            power = power/len(data)
            powerList.append(power)
    print(selectedFolders)
    return powerList, freq
```

```python
power_dict = {}
for folder in folders:
    power, freq = psd(f'data/{folder}', windows, channel, welchMethod)
    power_dict[folder] = {'power': power, 'freq': freq}
```

```python
fig = plt.figure(figsize=(10,5))
plt.rcParams.update({'font.size': 14})
plt.rcParams["axes.linewidth"] = 1

ax = plt.gca()
for key in power_dict.keys():
    ax.plot(power_dict[key]['freq'], power_dict[key]['power'][0],label = key)
#    ax.set_xlim([10, f/2])
#    ax.set_ylim([min(powerList[i][1:]), max(powerList[i])])
        
ax.set_yscale('log')
#ax.set_xscale('log')
ax.legend(['PSD - Without Filter', 'Laser Noise - without filter', 'Filter - without particle', 'Filter'])
ax.set(ylabel=r'$V^2$/Hz')
ax.set(xlabel=r'freq [Hz]')
ax.set_xlim([10000,120000])
ax.grid(alpha = 0.4)
```
