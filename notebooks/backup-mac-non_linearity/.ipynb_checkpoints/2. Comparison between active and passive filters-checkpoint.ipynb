```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df = pd.read_csv('Draft6.txt', sep='\t')
```

```python
df.columns
```

```python
df['mag'] = df['V(n002)'].apply(lambda x: float(x.split(',')[0][1:-2]))#values[3000]
```

```python
df['phase'] = df['V(n002)'].apply(lambda x: float(x.split(',')[1][:-2]))#values[3000]
```

```python
df['phase']
```

```python
plt.xscale('log')
plt.plot(df['Freq.'].values[1:], df['mag'].values[1:])
plt.
```

```python
fig, ax1 = plt.subplots()

plt.rcParams.update({'font.size': 18})
plt.rcParams["axes.linewidth"] = 2

color = 'tab:red'
ax1.set_xlabel('Freq (Hz)')
ax1.grid(True, which="both", ls="-")


ax1.set_ylabel('Mag (dB)')
ax1.plot(df['Freq.'], df['mag'], color=color)

ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Phase ()')
ax2.set_xscale('log')
ax2.plot(df['Freq.'], df['phase'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_xlim([1e0, 1e6])

fig.tight_layout()  # otherwise the right y-label is slightly clipped
#plt.grid()
#matplotlib.pyplot.grid(True, which="both")
plt.show()
```
