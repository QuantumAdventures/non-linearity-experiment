```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
b, a = signal.butter(4, 100, 'low', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20*np.log10(abs(h)))
plt.show()
```

```python
fs = 8e5
b, a = signal.butter(4, [10, 100], fs=fs, btype='band')
```

```python
fs = 800000.0
lowcut = 50000.0
highcut = 125000.0

# Plot the frequency response for a few different orders.
plt.figure(1)
plt.clf()
for order in [3, 6, 9]:
    b, a = signal.butter(order, [lowcut, highcut], fs=fs, btype='band', analog=False)
    w, h = signal.freqz(b, a, fs=fs, worN=2000)
    plt.plot(w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         '--', label='sqrt(0.5)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.grid(True)
plt.legend(loc='best')

# Filter a noisy signal.
T = 0.001
nsamples = T * fs
t = np.arange(0, nsamples) / fs
a = 0.02
f0 = 100000.0
x = 0.1 * np.sin(2 * np.pi * 1200 * np.sqrt(t))
x += 0.01 * np.cos(2 * np.pi * 3120 * t + 0.1)
x += a * np.cos(2 * np.pi * f0 * t + .11)
x += 0.03 * np.cos(2 * np.pi * 20000 * t)
plt.figure(2)
plt.clf()
plt.plot(t, x, label='Noisy signal')

b, a = signal.butter(3, [lowcut, highcut], fs=fs, btype='band', analog=False)
y = signal.lfilter(b, a, x)
print(y)
plt.plot(t, y, label='Filtered signal (%g Hz)' % f0)
plt.xlabel('time (seconds)')
#plt.hlines([-a, a], 0, T, linestyles='--')
plt.grid(True)
plt.axis('tight')
plt.legend(loc='upper left')

plt.show()
```

```python
b, a = signal.butter(3, [lowcut, highcut], fs=fs, btype='band', analog=False)
x_window = np.zeros(b.shape)
y_window = np.zeros((a.shape[0]-1))
y_filter = []
y_i = 0
for i in range(x.shape[0]):
    x_window[1:] = x_window[:-1]
    x_window[0]  = x[i]
    y_i = (b*x_window).sum()-(a[1:]*y_window).sum()
    y_window[1:] = y_window[:-1]
    y_window[0] = y_i


    y_filter.append(y_i)
```

```python
print(b.shape, a)
plt.plot(t, x)
plt.plot(t, y)
plt.plot(t, y_filter)
#plt.xlim([0.0000,0.0001])
#plt.ylim([-0.03, 0.03])
plt.show()
```
